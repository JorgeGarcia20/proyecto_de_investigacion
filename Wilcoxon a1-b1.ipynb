```python
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Experimentos/Datos/a1-b1.csv')
df1.head(10)
```

## **Perceptrón multicapa**

```python
# Perceptron multicapa
wilcoxon(df1['mlp-a1'], df1['mlp-b1'])
```

```python
ttest(df1['mlp-a1'], df1['mlp-b1'], equal_variances=False, paired=True)
```

## **Bosques aleatorios**

```python
wilcoxon(df1['rf-a1'], df1['rf-b1'])
```

```python
ttest(df1['rf-a1'], df1['rf-b1'], equal_variances=False, paired=True)
```
