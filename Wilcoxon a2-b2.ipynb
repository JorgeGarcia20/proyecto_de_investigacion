```python
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Experimentos/Datos/a2-b2.csv')
df1.head(10)
```

## **Perceptrón multicapa**

```python
# Perceptron multicapa
wilcoxon(df1['mlp-a2'], df1['mlp-b2'])
```

```python
ttest(df1['mlp-a2'], df1['mlp-b2'], equal_variances=False, paired=True)
```

## **Bosques aleatorios**

```python
wilcoxon(df1['rf-a2'], df1['rf-b2'])
```

```python
ttest(df1['rf-a2'], df1['rf-b2'], equal_variances=False, paired=True)
```
